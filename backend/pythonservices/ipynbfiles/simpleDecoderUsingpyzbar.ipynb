```python
import cv2
from pyzbar.pyzbar import decode
```

```python
fileInputPath = "../../sample-barcodeimages/dentastick.jpg"
```

```python
img = cv2.imread(fileInputPath)

barcodes = decode(img)

if barcodes:
    for i, barcode in enumerate(barcodes):
        x, y, w, h = barcode.rect
        cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 2)
        print(f"Barcode {i} data:", barcode.data.decode('utf-8'))
    cv2.imwrite("barcode_detected.png", img)
    print("Saved barcode_detected.png")
else:
    print("No barcode detected")
```
